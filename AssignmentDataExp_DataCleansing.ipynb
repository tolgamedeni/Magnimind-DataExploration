```python
#so for simplification, first category (1, high) will be over 900
#second category (2,normal) will be between 114 to 900 
# the third category(3,low) will be between 47 to 113
# the final category(4,very low) will be lower 46


def discretization(Revenue):
    if  Revenue >900:
        return 1
    elif Revenue > 113 :
        return 2
    elif Revenue >=47:
        return 3
    else: 
        return 4

    
    
movie_df['rank_discrete'] = movie_df['Rank'].apply(discretization)
movie_df.head(20)
```

```python
movie_df.groupby("rank_discrete")["Rank"].count()
```

```python
def discretization(Revenue_Cat):
    if  Revenue_Cat >900:
        return 'high'
    elif Revenue_Cat > 113 :
        return 'normal'
    elif Revenue_Cat >=47:
        return 'low'
    else: 
        return 'very low'

    
    
movie_df['rank_cat'] = movie_df['Rank'].apply(discretization)
movie_df.head(20)
```

```python
movie_df.groupby("rank_cat")["Rank"].count()
```
