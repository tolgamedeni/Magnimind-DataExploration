```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
edu_df=pd.read_csv("states_all.csv")


warnings.filterwarnings("ignore")
plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = (10,5)
#In this question to be able to see the outliers, first of all we p

#Question 01, to detect possbile outliers we apply the techniques we learned
#First Histogram
plt.hist(edu_df['TOTAL_REVENUE'])
plt.title("Total Reveunue")
plt.show()
```

```python
plt.hist(edu_df['TOTAL_EXPENDITURE'])
plt.title("Total Expenditure")
plt.show()
```

```python
#And finally, Z-Score
from scipy.stats import zscore
zscores = zscore(edu_df["TOTAL_REVENUE"])
for i in range(1,5):
    print("The score threshold is: {}".format(i))
    print("The indices of the outliers:")
    print(np.where(zscores > i))
    print("Number of outliers is: {}".format(len((np.where(zscores > i)[0]))))
    print('--'*40)
```

```python
from scipy.stats import zscore
zscores = zscore(edu_df["TOTAL_EXPENDITURE"])
for i in range(1,5):
    print("The score threshold is: {}".format(i))
    print("The indices of the outliers:")
    print(np.where(zscores > i))
    print("Number of outliers is: {}".format(len((np.where(zscores > i)[0]))))
    print('--'*40)
```

```python
#Question02
#We may apply the winsorize method, as you can see for revenue and expenditure, the outliers are on the upper limit 
#which means without lower limit, so:
from scipy.stats.mstats import winsorize
winsorizing= winsorize(np.array(edu_df['TOTAL_REVENUE'].dropna()), (0, 0.087))
winsorizing
```

```python
plt.boxplot(winsorizing)

plt.title("Total Revenue after winsorizing")

plt.show()
```

```python
#For the Total Revenue, we gave 8.7% as an upper limit, belove this level we were still taking outlier values
winsorizing= winsorize(np.array(edu_df['TOTAL_EXPENDITURE'].dropna()), (0, 0.088))
winsorizing
```

```python
plt.boxplot(winsorizing)

plt.title("Total Expenditure after winsorizing")

plt.show()
```

```python
#For the Total Expenditure, we gave 8.8% as an upper limit, belove this level we were still taking outlier values
```

```python
#Question 03 Budget_Deficit
edu_df['Budget_Deficit']=edu_df['TOTAL_REVENUE']-edu_df['TOTAL_EXPENDITURE']
```

```python
#To detect outliers, we will apply zscore with removing null values
```

```python
zscores = zscore(edu_df['Budget_Deficit'].dropna())
```

```python
for i in range(1,5):
    print("The score threshold is: {}".format(i))
    print("The indices of the outliers:")
    print(np.where(zscores > i))
    print("Number of outliers is: {}".format(len((np.where(zscores > i)[0]))))
    print('--'*40)
```

```python
#now for visualization:
plt.boxplot(edu_df['Budget_Deficit'].dropna())
plt.title("Budget Deficit")
plt.show()
            
```

```python
#As shown in the Budget Deficit figure, there are outlier values over and under upper limits so to eliminate them :
from scipy.stats.mstats import winsorize
winsorizing= winsorize(np.array(edu_df['Budget_Deficit'].dropna()), (0.0895, 0.067))
winsorizing
```

```python
plt.boxplot(winsorizing)

plt.title("Budget Deficit After Winsorizing")

plt.show()
```

```python
#As you can see, to eliminate the outlier values, for the lower limit 8.95% is implemented and for the upper limit 6.7% was
#implemented
#To eliminate the outlier values, the winsorize s more suitable
```
