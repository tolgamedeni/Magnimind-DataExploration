```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
from scipy.stats.mstats import winsorize
from sqlalchemy import create_engine
import warnings
warnings.filterwarnings('ignore')
sns.set(style="whitegrid")
pd.options.display.float_format = '{:.2f}'.format
plt.rcParams['figure.dpi'] = 100
student_df = pd.read_csv('StudentsPerformance.csv')
```

```python
#As you can see, we only get, correlation values from the numerical columns. In follow, we will try to group categorical columns with numerical ones
#Lets start with math and apply for gender, origin and education
student_df.groupby("gender")['math score'].mean().T
```

```python
#Gender
students=student_df["gender"].unique()
student_df.dropna(inplace=True)
pd.options.display.float_format = '{:.4f}'.format

for var in ["math score", "reading score", "writing score"]:
    comparison = pd.DataFrame(columns=['group_1', 'group_2','statistic', 'p_value'])
    print("{}".format(var),end='')
    for i in range(0, len(students)):
        for j in range(i+1, len(students)):
            ttest = stats.ttest_ind(student_df[student_df["gender"]==students[i]][var],student_df[student_df["gender"]==students[j]][var]) 
            group_1 = students[i]
            group_2 = students[j]
            statistic = ttest[0]
            p_value = ttest[1]
            comparison = comparison.append({"group_1" : group_1 ,"group_2" : group_2 , "statistic": statistic ,  "p_value" : p_value}, ignore_index=True)
    display(comparison)


```

```python
#we need to reject null hyp. which is gender has effect on the score. So the result is in favor of alternative hypothesis this indicates that gender makes no difference on scores.
#So gender has no significant difference on the scores.

#Let's continue with origin



students=student_df["race/ethnicity"].unique()
student_df.dropna(inplace=True)
pd.options.display.float_format = '{:.4f}'.format

for var in ["math score", "reading score", "writing score"]:
    comparison = pd.DataFrame(columns=['group_1', 'group_2','statistic', 'p_value'])
    print("{}".format(var),end='')
    for i in range(0, len(students)):
        for j in range(i+1, len(students)):
            ttest = stats.ttest_ind(student_df[student_df["race/ethnicity"]==students[i]][var],student_df[student_df["race/ethnicity"]==students[j]][var]) 
            group_1 = students[i]
            group_2 = students[j]
            statistic = ttest[0]
            p_value = ttest[1]
            comparison = comparison.append({"group_1" : group_1 ,"group_2" : group_2 , "statistic": statistic ,  "p_value" : p_value}, ignore_index=True)
    display(comparison)


```

```python
#As it has been seen on math score, reading and writing scores for p_values which are greater than 5%(0.05), the origin makes difference
#Such as on math,reading and writing score, groupB and groupC with the p_value of 0.464, 0.1867 and 0.1110 shows accepting the null
#hypothesis. However for the group B and group D, the values area 0.000, 0.0008 and 0.0008 so null hypothesis does not acceptable
```
